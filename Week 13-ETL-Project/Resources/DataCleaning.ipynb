```python
missing_values = ["n/a", "none", "NaN", "--"]
listings_df = pd.read_csv('seattle_airbb\listings.csv', na_values = missing_values)
listings_df.head(2)
```

```python
print (listings_df['experiences_offered'].isnull())
```

```python
cnt=0
for row in listings_df['experiences_offered']:
    try:
        int(row)
        listings_df.loc[cnt, 'experiences_offered']=np.nan
    except ValueError:
        pass
    cnt+=1
```
